```python
#Importo
df = pd.read_excel('C:/Users/Usuario/Desktop/Efecty/Proyecto VOZ/CRM - Dinamycs Encuestas/Respuestas totales encuestas papa para entrenamiento.xlsx')
# Vista de búsqueda avanzada de casos 29-09-2021 10-33-48.xlsx')

# 
#Encuestas=Encuestas[(Encuestas['Canal de ingreso']=='Web')|(Encuestas['Canal de ingreso']=='Correo electrónico')]
df.reset_index(inplace=True)
df.rename(columns={'index':'index_inicial'},inplace=True)

df=df.iloc[0:800,:]# Encuestas['Descripción']
```

```python
# df[['comentario','category']]

df['len_comentario']=df['comentario'].apply(lambda x:len(str(x)))
df['len_comentario']=df['len_comentario'].apply(lambda x:120 if x>30 else x)
   
```

```python
#TRAINNING_DATA
TRAINNING_DATA=[]
for j in range(0,len(df)):
    TRAINNING_DATA.append((str(df['comentario'][j]),{'entities':[(0,int(df['len_comentario'][j]),str(df['category'][j]))]}))
```

```python
from spacy.training.example import Example

for batch in spacy.util.minibatch(TRAINNING_DATA, size=2):
    for text, annotations in batch:
        # create Example
        doc = nlp.make_doc(text)
        example = Example.from_dict(doc, annotations)
        # Update the model
        nlp.update([example], drop=0.3)
```
